```python
from numpy.random import default_rng
rng = default_rng()
mu = 50000
std = 5000
data = rng.normal(loc=mu, scale=std, size=1000)
print(np.mean(data), np.std(data))

fig, ax = plt.subplots(figsize=(8,4))
ax.hist(data, bins=150, density=True)
ax.set_ylabel('P')
ax.set_xlabel('Price')
ax.tick_params(axis='x', labelrotation=45)
ax.xaxis.set_major_formatter('€ {x:,.0f}')
plt.show()
```

```python
build = rng.normal(loc=50000, scale=5000, size=500)
land = rng.normal(loc=10000, scale=500, size=500)

fig, ax = plt.subplots(figsize=(8,4))
result = np.array([np.random.choice(build) + np.random.choice(land) for _ in range(1000000)])
ax.hist(result, bins=150, density=True)
ax.set_ylabel('P')
ax.set_xlabel('Price')
ax.tick_params(axis='x', labelrotation=45)
ax.xaxis.set_major_formatter('€ {x:,.0f}')
plt.show()
```
